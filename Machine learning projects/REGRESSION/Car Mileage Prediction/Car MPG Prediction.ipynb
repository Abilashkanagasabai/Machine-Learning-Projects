# PREPROCESSING THE DATA
     *MISSING VALUES
     *CATEGORIAL VALUES

```python
df_obj=df.dtypes[df.dtypes=="object"].index
```

```python
df[df_obj]
```

```python
temp=pd.DataFrame(df.hp.str.isdigit())
temp[temp['hp']==False]
```

```python
temp[temp["hp"]==False]
```

```python
df=df.replace("?",np.nan)
```

```python
sns.heatmap(df.isnull())
```

```python
df['hp'].fillna((df['hp'].median()),inplace=True)
df.hp.median()
```

```python
df.dtypes
```

```python
df.hp=df.hp.astype("float64")
```

```python
df.dtypes
```

```python
df["car name"].value_counts()
```

```python
df.drop("car name",axis=1,inplace=True)
```

# Replacing the origin code to countries

```python
df["origin"]=df["origin"].replace({1:'america',2:'europe',3:'asia'})
df
```

# ONE HOT ENCODING

```python
df = pd.get_dummies(df, columns=['origin'])
df
```

# SPLITTING THE DEPENDENT AND INDEPENDENT VARIABLES

```python
X=df.drop("mpg",axis=1)
y=df[["mpg"]]
X.head()
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
#MODEL EVALUATION
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

# RIDGE AND LASSO REGRESSION (L1&L2 REGULARIZATION)

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40]}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)

lasso_regressor.fit(X_train,y_train)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
y_pred_lasso=lasso_regressor.predict(X_test)
```

```python
#MODEL EVALUATION

print('MAE:', metrics.mean_absolute_error(y_test, y_pred_lasso))
print('MSE:', metrics.mean_squared_error(y_test, y_pred_lasso))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_lasso)))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred_lasso)
```

# K NEAREST NEIGHBOUR 

```python
from sklearn.neighbors import KNeighborsRegressor
knn_reg = KNeighborsRegressor(n_neighbors=1)
knn_reg.fit(X_train_scaled, y_train_scaled)
```

```python
y_pred_knn=knn_reg.predict(X_test_scaled)
```

```python
error = []
for i in range(1,49):  
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(X_train_scaled, y_train_scaled)
    pred_i = knn.predict(X_test_scaled)
    error.append(np.mean(pred_i != y_test_scaled))
plt.figure(figsize=(12, 6))  
plt.plot(range(1,49), error, color='red', linestyle='dashed', marker='o',  
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')  
plt.xlabel('K Value')  
plt.ylabel('Mean Error') 
```

```python
#MODEL EVALUATION

print('MAE:', metrics.mean_absolute_error(y_test_scaled, y_pred_knn))
print('MSE:', metrics.mean_squared_error(y_test_scaled, y_pred_knn))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test_scaled, y_pred_knn)))
from sklearn.metrics import r2_score
r2_score(y_test_scaled, y_pred_knn)
```
