```python
def outliers(var):
    a = []
    q1 = df[var].quantile(.25)
    q2 = df[var].quantile(.5)
    q3 = df[var].quantile(.75)
    iqr = q3-q1
    ulim = float(q3+(1.5*iqr))
    llim = float(q1-(1.5*iqr))

    for i in df[var]:
        if i > ulim:
            i=np.NaN
        elif i < llim:
            i = np.NaN
        else:
            i=i
        a.append(i)
    return a

for col in df.select_dtypes(exclude='object').columns:
    df[col] = outliers(col)
```

```python
df1=pd.get_dummies(df.cut,drop_first=True)
df=pd.concat([df,df1],axis=1)
del df["cut"]
df.head()
```

```python
df2=pd.get_dummies(df.color,drop_first=True)
df=pd.concat([df,df2],axis=1)
del df["color"]
df.head()
```

```python
df3=pd.get_dummies(df.clarity,drop_first=True)
df=pd.concat([df,df3],axis=1)
del df["clarity"]
df.head()
```

```python
X=df.loc[:,df.columns!="price"]
y=df.iloc[:,3]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
print(X_train.shape,"X_train data shape")
print(X_test.shape,"X_test data shape")
print(y_train.shape,"y_train data shape")
print(y_test.shape,"y_test data shape")
```

```python
import xgboost
from xgboost import XGBRegressor
lr=XGBRegressor()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
y_pred_train=lr.predict(X_train)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
print(r2_score(y_train,y_pred_train))
```

```python
from sklearn import metrics 

from sklearn.ensemble import RandomForestRegressor
rfregressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
rfregressor.fit(X_train,y_train)
y_pred_randomforest = rfregressor.predict(X_test)
#MODEL EVALUATION

print('MAE:', metrics.mean_absolute_error(y_test, y_pred_randomforest))
print('MSE:', metrics.mean_squared_error(y_test, y_pred_randomforest))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_randomforest)))
```
